# Scatterplot (gráfico de dispersión)

## ¿Qué es un Gráfico de Dispersión?

Un gráfico de dispersión, también conocido como scatter plot, es una representación gráfica de datos donde cada punto en el gráfico representa un par de valores de dos variables diferentes. En el eje horizontal (X) se coloca una variable, y en el eje vertical (Y) se coloca la otra variable. Cada punto en el gráfico muestra cómo se relacionan los valores de estas dos variables.

## ¿Para qué se Usa?

Los gráficos de dispersión se utilizan principalmente para:

1. **Identificar Relaciones**:
   - Ayudan a visualizar la relación entre dos variables, permitiendo identificar si existe una correlación positiva, negativa o ninguna correlación entre ellas.
   - Por ejemplo, un gráfico de dispersión puede mostrar la relación entre la altura y el peso de una muestra de personas.

2. **Detectar Tendencias**:
   - Al observar el patrón de los puntos, se pueden detectar tendencias o patrones en los datos.
   - Por ejemplo, si los puntos siguen una línea ascendente, indica que a medida que una variable aumenta, la otra también lo hace.

3. **Identificar Outliers**:
   - Los gráficos de dispersión pueden ayudar a identificar outliers, que son puntos de datos que se desvían significativamente del resto de los puntos.
   - Estos outliers pueden representar errores de medición, datos anómalos o eventos raros que merecen una atención especial.

4. **Evaluar la Dispersión de los Datos**:
   - Permiten evaluar cómo se dispersan los datos en el espacio, es decir, si los puntos están muy juntos o dispersos.
   - Esto ayuda a entender la variabilidad de los datos.

## Ejemplo de Gráfico de Dispersión

Imaginemos que queremos estudiar la relación entre las horas de estudio y las calificaciones de los estudiantes en un examen. Podemos recopilar datos de varios estudiantes y representarlos en un gráfico de dispersión:

- En el eje X colocamos las horas de estudio.
- En el eje Y colocamos las calificaciones obtenidas.

Cada punto en el gráfico representará un estudiante, mostrando cuántas horas estudió y qué calificación obtuvo. Al observar el gráfico, podemos identificar si existe una tendencia que indique que más horas de estudio se relacionan con mejores calificaciones.

## Variables Dependiente e Independiente

En algunos análisis de datos, es importante distinguir entre dos tipos de variables: dependiente e independiente. Sin embargo, no todos los análisis requieren esta distinción. Específicamente, es crucial hacer esta diferenciación cuando estamos interesados en entender cómo una variable afecta a otra.

### Cuándo Hacer la Distinción

Esta distinción es necesaria en estudios experimentales y de causalidad donde se desea observar el efecto de una variable sobre otra. Por ejemplo:

- **Estudios de Causalidad**: Queremos saber cómo el tiempo de estudio (variable independiente) afecta las calificaciones (variable dependiente).
- **Experimentos Controlados**: Estamos interesados en ver cómo diferentes dosis de un medicamento (variable independiente) afectan la salud de los pacientes (variable dependiente).

En estos casos, la identificación clara de variables independientes y dependientes permite diseñar experimentos y análisis que pueden establecer relaciones causales.


### Variable Independiente

- **Definición**: La variable independiente es aquella que se manipula o controla en un experimento para observar su efecto en la variable dependiente.
- **Ubicación en el Gráfico**: Por convención, la variable independiente se coloca en el eje horizontal (X).

### Variable Dependiente

- **Definición**: La variable dependiente es la que se observa y se mide para ver cómo cambia en respuesta a la variable independiente.
- **Ubicación en el Gráfico**: Por convención, la variable dependiente se coloca en el eje vertical (Y).

### Ejemplo

Imaginemos que estamos estudiando cómo el tiempo de estudio afecta las calificaciones de los estudiantes:

- **Variable Independiente**: Tiempo de estudio (colocado en el eje X).
- **Variable Dependiente**: Calificaciones obtenidas (colocadas en el eje Y).

En este contexto, estamos interesados en ver cómo los cambios en el tiempo de estudio (variable independiente) influyen en las calificaciones (variable dependiente).


```python
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados: horas de estudio y calificaciones
horas_estudio = np.random.rand(100) * 10  # Variable independiente
calificaciones = 50 + horas_estudio * 5 + np.random.randn(100) * 5  # Variable dependiente

# Crear scatter plot
plt.scatter(horas_estudio, calificaciones, color='blue')
plt.title("Relación entre Horas de Estudio y Calificaciones")
plt.xlabel("Horas de Estudio (Variable Independiente)")
plt.ylabel("Calificaciones (Variable Dependiente)")
plt.show()
```

# Cómo los Outliers Afectan Nuestros Scatterplots
Los outliers son observaciones que se desvían significativamente de otros puntos de datos en un conjunto de datos. Pueden deberse a variabilidad en los datos o a errores experimentales. La detección y el manejo de outliers son cruciales para el análisis de datos, ya que pueden influir en las conclusiones y modelos derivados de los datos.
 

# Impacto de los Outliers en Scatterplots
1. Distorsión de la Relación Entre Variables
Los outliers pueden distorsionar la relación aparente entre dos variables en un scatterplot. Esto puede llevar a interpretaciones incorrectas sobre la fuerza y la dirección de la relación entre las variables. Por ejemplo, en un scatterplot de peso versus altura, unos pocos puntos de datos extremadamente altos o bajos pueden sesgar la percepción de la correlación entre estas dos variables.

2. Afectación de la Pendiente y la Intersección en la Regresión
En la regresión lineal, los outliers pueden tener un impacto significativo en la pendiente y la intersección de la línea de mejor ajuste. Esto se debe a que la regresión lineal trata de minimizar la suma de los cuadrados de las distancias (residuales) de todos los puntos a la línea de mejor ajuste. Los outliers, al estar alejados de la mayoría de los datos, pueden aumentar considerablemente estos residuales, alterando la línea de regresión.

3. Incremento de la Dispersión Aparente
La presencia de outliers puede aumentar la dispersión aparente de los datos en un scatterplot. Esto puede hacer que los datos parezcan más variables de lo que realmente son, afectando medidas de dispersión como la varianza y la desviación estándar.

4. Sesgo en la Identificación de Patrones
Los outliers pueden crear patrones espurios que no representan la realidad de los datos subyacentes. Estos patrones pueden llevar a decisiones erróneas en análisis posteriores, como la identificación de clusters o la segmentación de datos.

5. Afectación de las Métricas de Evaluación
Las métricas utilizadas para evaluar la calidad de modelos predictivos, como el error cuadrático medio (MSE) o el coeficiente de determinación (R²), pueden ser severamente afectadas por outliers. Esto puede dar una falsa impresión de la precisión y la eficacia de un modelo.


# Técnicas de Manejo de Outliers

## Detección Estadística

### Rango Intercuartílico (IQR)
El rango intercuartílico es una medida estadística utilizada para identificar outliers. <br>Se calcula como la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1). 

En un conjunto de datos, un valor puede ser considerado un valor atípico si está por debajo de $ (Q1 - 1.5 \times IQR) $ o por encima de $ (Q3 + 1.5 \times IQR) $.

$$
\text{IQR} = Q3 - Q1
$$

- **Límite Inferior:** 
  $$
  Q1 - 1.5 \times \text{IQR}
  $$
- **Límite Superior:** 
  $$
  Q3 + 1.5 \times \text{IQR}
  $$

### Desviación Estándar
La desviación estándar mide la dispersión de un conjunto de datos respecto a su media. Se considera que los puntos de datos que se desvían más allá de un cierto número de desviaciones estándar (típicamente 2 o 3) de la media son outliers.

## Métodos de Machine Learning

### Clustering
Las técnicas de clustering agrupan los datos en clusters o grupos. Los puntos de datos que no pertenecen a ningún cluster principal se consideran outliers. Ejemplos de algoritmos de clustering incluyen K-means y DBSCAN.

### Modelos de Detección de Anomalías
Se emplean modelos específicos para la detección de anomalías, que identifican outliers basándose en patrones aprendidos de los datos. Algunos ejemplos de estos modelos son:
- **Isolation Forest:** Un algoritmo basado en árboles que detecta outliers dividiendo aleatoriamente los datos.
- **Local Outlier Factor (LOF):** Mide la densidad local de un punto en comparación con sus vecinos, identificando outliers como aquellos puntos con menor densidad.

Estas técnicas combinadas permiten un enfoque robusto para la identificación y el manejo de outliers, mejorando así la calidad de los análisis y modelos predictivos.

# Correlación

## ¿Por Qué se Usa el Gráfico de Dispersión para la Correlación?

El gráfico de dispersión es una herramienta útil para analizar la correlación porque permite visualizar directamente la relación entre dos variables. Al observar cómo se distribuyen los puntos en el gráfico, se puede identificar si las variables están relacionadas y de qué manera:

1. **Correlación Positiva**:
   - Cuando los puntos en el gráfico tienden a seguir una línea ascendente, se dice que hay una correlación positiva. Esto significa que a medida que una variable aumenta, la otra también lo hace.
   
2. **Correlación Negativa**:
   - Cuando los puntos tienden a seguir una línea descendente, hay una correlación negativa. Esto indica que a medida que una variable aumenta, la otra disminuye.

3. **No Correlación**:
   - Si los puntos están dispersos sin seguir un patrón claro, no hay correlación aparente entre las variables.


# Coeficiente de Correlación de Pearson y su Relación con los Scatterplots

## ¿Qué es el Coeficiente de Correlación de Pearson?

El coeficiente de correlación de Pearson, denotado como $r$, es una medida estadística que indica la fuerza y la dirección de la relación lineal entre dos variables cuantitativas. Se calcula utilizando la siguiente fórmula:

$ r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}} \ $

- $x_i$ y $y_i$ son los valores individuales de las variables $x$ y $y$.
- $\bar{x}$ y $\bar{y}$ son las medias de las variables $x$ y $y$.

El valor de $r$ varía entre -1 y 1:
- **$r = 1$**: Correlación positiva perfecta.
- **$r = -1$**: Correlación negativa perfecta.
- **$r = 0$**: No hay correlación lineal.


## Limitaciones del Coeficiente de Correlación de Pearson

Es importante destacar que el coeficiente de correlación de Pearson solo mide relaciones lineales entre variables. No puede detectar relaciones no lineales. Existen otros tipos de correlaciones para diferentes relaciones, como la correlación de Spearman, que es adecuada para datos que no siguen una distribución normal o que tienen una relación monótona pero no necesariamente lineal.

## Distinción entre Variables

### Variables Numéricas

Las variables numéricas son aquellas que representan cantidades y pueden ser medidas y ordenadas. Ejemplos incluyen horas de estudio, calificaciones, altura y peso. La correlación de Pearson es adecuada para medir la relación lineal entre dos variables numéricas.

### Variables Categóricas

Las variables categóricas representan categorías o grupos. Ejemplos incluyen género, estado civil, tipo de producto, etc. Para estas variables, se utilizan otras técnicas de análisis, como:
- **Chi-cuadrado (Chi²)**: Utilizado para evaluar la asociación entre dos variables categóricas.
- **ANOVA (Análisis de Varianza)**: Utilizado para comparar las medias de tres o más grupos categóricos.

### Series de Tiempo

En el análisis de series de tiempo, la variable de fecha o tiempo cambia de forma natural y secuencial. La correlación de Pearson no es adecuada para series de tiempo porque el tiempo es una variable que cambia por sí misma y no de manera dependiente. En lugar de esto, se utilizan técnicas como:
- **Autocorrelación**: Mide la correlación de una serie de tiempo consigo misma en diferentes momentos.
- **Correlación cruzada**: Mide la correlación entre dos series de tiempo.
